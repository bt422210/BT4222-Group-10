##2.1 Popularity

*   Click Through Rate of each news article
*   Calculated by (total click count for article / appearance count in impression)
*   To prevent data leakage, Click Through Rate is calculated from training data only
*   Unseen articles in the validation and test sets are assigned a default CTR value equal to the overall training CTR
*   To reduce skewness in data, CTR is subsequently log transformed and normalized as `ctr_norm`



```python
import ast

train.columns = train.columns.str.strip()

# 1. Positive click counts
pos_counts = (
    train.groupby("pos_id", as_index=False)
    .size()
    .rename(columns={"pos_id": "news_id", "size": "total_clicks"})
)

# 2. Total impression counts (pos_id + neg_ids)
if train["neg_ids"].dtype == object:
    train["neg_ids"] = train["neg_ids"].apply(
        lambda x: ast.literal_eval(x) if isinstance(x, str) else x
    )
train_ctr = train.copy()

train_ctr["all_news_ids"] = train_ctr.apply(
    lambda row: [row["pos_id"]] + row["neg_ids"],
    axis=1
)

all_impressions = (
    train_ctr[["impression_id", "all_news_ids"]]
    .explode("all_news_ids")
    .rename(columns={"all_news_ids": "news_id"})
)

impression_counts = (
    all_impressions.groupby("news_id", as_index=False)
    .size()
    .rename(columns={"size": "total_impressions"})
)

# 3. Compute CTR
article_stats_train = impression_counts.merge(
    pos_counts,
    on="news_id",
    how="left"
)

article_stats_train["total_clicks"] = article_stats_train["total_clicks"].fillna(0)

article_stats_train["ctr"] = (
    article_stats_train["total_clicks"] / article_stats_train["total_impressions"]
)

# 4. Fallback CTR
total_clicks_all = len(train)  # 1 clicked article per impression
total_impressions_all = len(all_impressions)

overall_train_ctr = total_clicks_all / total_impressions_all

print("Summary Statistics for Train-derived CTR")
print(article_stats_train["ctr"].describe())
print("--------------------------------------------------------")
print(f"Overall train CTR (fallback) = {overall_train_ctr:.6f}")

# 5. Log transform CTR
article_stats_train["ctr_log"] = np.log1p(article_stats_train["ctr"])

print("\nSummary Statistics for Train-derived log CTR")
print(article_stats_train["ctr_log"].describe())

# 6. Merge into news_features
news_features.drop(
    columns=["ctr", "ctr_log", "ctr_norm", "ctr_x", "ctr_y"],
    errors="ignore",
    inplace=True
)

news_features = news_features.merge(
    article_stats_train[["news_id", "ctr", "ctr_log"]],
    on="news_id",
    how="left"
)

num_unseen_articles = news_features["ctr"].isna().sum()

# fill unseen articles with global fallback CTR
news_features["ctr"] = news_features["ctr"].fillna(overall_train_ctr)
news_features["ctr_log"] = news_features["ctr_log"].fillna(np.log1p(overall_train_ctr))

# 7. Normalize log CTR using TRAIN-only stats
ctr_log_mean = article_stats_train["ctr_log"].mean()
ctr_log_std = article_stats_train["ctr_log"].std()

if pd.isna(ctr_log_std) or ctr_log_std == 0:
    ctr_log_std = 1.0

news_features["ctr_norm"] = (
    (news_features["ctr_log"] - ctr_log_mean) / ctr_log_std
)

# 8. Checks
print(f"\nNumber of unseen articles filled with overall train CTR = {num_unseen_articles}")

print("\nRaw CTR summary in news_features:")
print(news_features["ctr"].describe())

print("\nLog CTR summary in news_features:")
print(news_features["ctr_log"].describe())

print("\nNormalized log CTR summary in news_features:")
print(news_features["ctr_norm"].describe())

print("\nnews_features DataFrame after merging CTR features:")
print(news_features[["news_id", "ctr", "ctr_log", "ctr_norm"]].head())

news_features.drop(columns=["ctr", "ctr_log"], inplace=True, errors="ignore")
```

##2.3 Latent Topics

- Latent topics are derived by applying Singular Value Decomposition (SVD) to the TF-IDF representations of all news articles.
- This step compresses the high‑dimensional TF‑IDF vectors to the top 50 components, corresponding to the most informative latent topics in the corpus.
- Each article is then assigned a 50‑dimensional topic score vector, indicating how strongly it aligns with each latent theme. This provides a compact, interpretable semantic representation that complements the other content‑based features.

- Latent topics are then merged into the `news_features` dataframe



```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import pandas as pd

# Ensure the 'title' column has no NaN values, fill with empty strings if any
news['title'] = news['title'].fillna('')

tfidf = TfidfVectorizer(
    stop_words='english',   # Removes common words like 'the', 'is', etc.
    max_features=1000,      # Keeps only the top 1000 most important words
    ngram_range=(1, 2)      # Captures both single words and 2-word phrases (e.g., "Artificial Intelligence")
)

tfidf_matrix = tfidf.fit_transform(news['title'])

# You can still create this df for inspection if needed, but it's not directly used for news features
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out(),
    index=news['news_id'] # Use news['news_id'] as the index
)

svd = TruncatedSVD(n_components=50, random_state=42)
tfidf_reduced = svd.fit_transform(tfidf_matrix)

# This 'tfidf_reduced' is now a much smaller, dense vector you can feed into your model

tfidf_reduced_df = pd.DataFrame(tfidf_reduced, index=news.index)
svd_stats = tfidf_reduced_df.describe().T
summary = tfidf_reduced_df.agg(['min', 'max', 'mean', 'median', 'std']).T
print("Summary Statistics for first 10 SVD Components:")
print(summary.head(10))

# Create a new DataFrame for latent topics
topic_cols = [f'topic_{i}' for i in range(tfidf_reduced.shape[1])]
news_topics_df = pd.DataFrame(tfidf_reduced, columns=topic_cols, index=news.index)
news_topics_df['news_id'] = news['news_id']

# Reorder columns to have 'news_id' first
news_topics_df = news_topics_df[['news_id'] + topic_cols]

print("\nNews Latent Topics DataFrame:")
print(news_topics_df.head())
```

```python
news_features = news_features.merge(
    news_topics_df,
    on='news_id',
    how='left'
)

print("News features after adding latent topics:")
print(news_features.head())
```

##2.4 First Train Impression Time (used for later features)
Serves as a time reference to compute Recency Score of news article for model input. *(See Section 5.3)*

* Computed using the training set only to prevent data leakage. Unseen articles are stored as NaN to be handled during recency computation (context feature).
* It is also used to weight the interest clustering (user feature)



```python
# 1. Ensure datetime
train = train.copy()
news = news.copy()

train["time"] = pd.to_datetime(train["time"])

# 2. Build article impressions (TRAIN only)
train_tmp = train.copy()

train_tmp["all_news_ids"] = train_tmp.apply(
    lambda row: [row["pos_id"]] + row["neg_ids"],
    axis=1
)

# explode → one row per article impression
train_article_impressions = (
    train_tmp[["time", "all_news_ids"]]
    .explode("all_news_ids")
    .rename(columns={"all_news_ids": "news_id"})
)

# 3. Compute first impression time
first_impression_lookup = (
    train_article_impressions
    .groupby("news_id", as_index=False)["time"]
    .min()
    .rename(columns={"time": "first_train_impression_time"})
)

print(first_impression_lookup.head())

# 4. Merge into news dataset
news = news.drop(columns=["first_train_impression_time"], errors="ignore")

news = news.merge(
    first_impression_lookup,
    on="news_id",
    how="left"
)

print(news[["news_id", "first_train_impression_time"]].head())

# 5. Check unseen articles
num_missing = news["first_train_impression_time"].isna().sum()
print(f"Number of articles not seen in training = {num_missing}")
```

## 3.1 Top 3 Favourite Categories
* Extracting the top 3 favourite categories for each user, depending on their historical data of clicked news articles
* Result in the addition of columns `top_cat_1`, `top_cat_2` and `top_cat_3`.


```python
from collections import Counter

news_cat_map = dict(zip(news['news_id'], news['subclustered_category']))

# extraction function
def get_top_3_categories(history_list):
    # Handle cold-start users (empty history)
    if pd.isna(history_list) is True:
        return ['cold_start', 'cold_start', 'cold_start']

    # Map IDs to categories using the dictionary lookup
    categories = [news_cat_map.get(nid) for nid in history_list if news_cat_map.get(nid) is not None]

    # Handle cases where none of the IDs had valid categories
    if not categories:
        return ['cold_start', 'cold_start', 'cold_start']

    # Count the frequencies of each category
    cat_counts = Counter(categories)

    # Extract the top 3 most common categories
    top_cats = [cat for cat, count in cat_counts.most_common(3)]

    # Pad with 'cold_start' if the user read fewer than 3 unique categories
    while len(top_cats) < 3:
        top_cats.append('cold_start')

    return top_cats

# Drop existing top_cat columns if they exist
train_user_features.drop(columns=['top_cat_1', 'top_cat_2', 'top_cat_3'], errors='ignore', inplace=True)

# 3. Apply the function to the History column
# This returns a pandas Series where each element is a list of 3 strings
top_3_series_train = train_user_features['history'].apply(get_top_3_categories)

# 4. Expand the lists into 3 distinct columns and assign them to the dataframe
train_user_features[['top_cat_1', 'top_cat_2', 'top_cat_3']] = pd.DataFrame(top_3_series_train.tolist(), index=top_3_series_train.index)

# Check the results
print(train_user_features[['history', 'top_cat_1', 'top_cat_2', 'top_cat_3']].head())
```

```python
# Apply to val_user_features
val_user_features.drop(columns=['top_cat_1', 'top_cat_2', 'top_cat_3'], errors='ignore', inplace=True)
top_3_series_val = val_user_features['history'].apply(get_top_3_categories)
val_user_features[['top_cat_1', 'top_cat_2', 'top_cat_3']] = pd.DataFrame(top_3_series_val.tolist(), index=top_3_series_val.index)

print("Validation user features after adding top 3 categories:")
print(val_user_features[['history', 'top_cat_1', 'top_cat_2', 'top_cat_3']].head())
```

```python
# Apply to test_user_features
test_user_features.drop(columns=['top_cat_1', 'top_cat_2', 'top_cat_3'], errors='ignore', inplace=True)
top_3_series_test = test_user_features['history'].apply(get_top_3_categories)
test_user_features[['top_cat_1', 'top_cat_2', 'top_cat_3']] = pd.DataFrame(top_3_series_test.tolist(), index=top_3_series_test.index)

print("Test user features after adding top 3 categories:")
print(test_user_features[['history', 'top_cat_1', 'top_cat_2', 'top_cat_3']].head())
```

## 3.2 Recent Favourite Category
* The most frequent news category extracted from the user's 10 most recent historical articles.
* For users with histories shorter than 10 items, all available records are evaluated. 'cold_start' is imputed for users with no prior history.



```python
from collections import Counter

# 1. Reuse the news_cat_map we created earlier
news_cat_map = dict(zip(news['news_id'], news['subclustered_category']))

# 2. Define the Recent Favourite Category function
def get_recent_fav_category(history_list_of_ints):
    # Handle users with no prior history (cold-start) or empty history list
    if pd.isna(history_list_of_ints) is True:
        return 'cold_start'

    # Take ONLY the 10 most recent items (the end of the list)
    recent_news_ids = history_list_of_ints[-10:]

    # Map those 10 IDs to their respective categories
    # news_cat_map keys are integers, and recent_news_ids contains integers
    categories = [news_cat_map.get(nid) for nid in recent_news_ids if news_cat_map.get(nid) is not None]

    # If no valid categories are found in the last 10 items
    if not categories:
        return 'cold_start'

    # Count frequencies and return the single most common category
    cat_counts = Counter(categories)
    return cat_counts.most_common(1)[0][0]

# 3. Apply the function to the metadata table
train_user_features['recent_fav_cat'] = train_user_features['history'].apply(get_recent_fav_category)

# 4. Verification check
print("\nDistribution of Recent Favourite Categories for train_user_features:")
print(train_user_features['recent_fav_cat'].value_counts().head(10))
```

##3.4 Diversity Cluster


* Used K-means clustering to segment user base into 4 distinct behaviour types (Cold Start, Low Diversity, Medium Diversity and High Diversity)
* Captures whether user reads broadly across different news categories or if they focus on a few specific categories only


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
import seaborn as sns
import collections
```

```python
train_user_features.drop(columns=['diversity_ratio', 'concentration_ratio'], errors='ignore', inplace=True)
val_user_features.drop(columns=['diversity_ratio', 'concentration_ratio'], errors='ignore', inplace=True)
test_user_features.drop(columns=['diversity_ratio', 'concentration_ratio'], errors='ignore', inplace=True)
```

```python
news_map = dict(zip(news['news_id'], news['category']))

def calculate_cluster_features(history_val):
 # 1. Standardize input: Convert strings/arrays/lists into a single list of IDs
    history_list = []

    if isinstance(history_val, str) and history_val.strip() != "":
        history_list = history_val.split()
    elif hasattr(history_val, '__len__') and not pd.isna(history_val).all():
        # This handles cases where history is already a list or numpy array
        history_list = list(history_val)

    # Handle cold-start or empty cases
    if not history_list:
        return [0.0, 0.0]

    # 2. Map IDs to categories using your news_map
    categories = [news_map.get(nid) for nid in history_list if nid in news_map]

    if not categories:
        return [0.0, 0.0]

    total_clicks = len(categories)
    unique_cats = len(set(categories))

    # 3. Calculate Diversity Ratio (Higher = user reads many different topics)
    diversity = unique_cats / total_clicks

    # 4. Calculate Concentration Ratio (Higher = user sticks to one main topic)
    counts = collections.Counter(categories)
    top_cat_count = counts.most_common(1)[0][1]
    concentration = top_cat_count / total_clicks

    return [diversity, concentration]

# Apply the function to train_user_features
cluster_features_train = train_user_features['history'].apply(calculate_cluster_features)

# Convert the resulting Series of lists into a DataFrame with named columns
train_features_df = pd.DataFrame(cluster_features_train.tolist(),
                           index=train_user_features.index,
                           columns=['diversity_ratio', 'concentration_ratio'])

# Add the columns
train_user_features = pd.concat([train_user_features, train_features_df], axis=1)

# Apply the function to val_user_features
cluster_features_val = val_user_features['history'].apply(calculate_cluster_features)

# Convert the resulting Series of lists into a DataFrame with named columns
val_features_df = pd.DataFrame(cluster_features_val.tolist(),
                           index=val_user_features.index,
                           columns=['diversity_ratio', 'concentration_ratio'])

# Add the columns
val_user_features = pd.concat([val_user_features, val_features_df], axis=1)

# Apply the function to test_user_features
cluster_features_test = test_user_features['history'].apply(calculate_cluster_features)

# Convert the resulting Series of lists into a DataFrame with named columns
test_features_df = pd.DataFrame(cluster_features_test.tolist(),
                           index=test_user_features.index,
                           columns=['diversity_ratio', 'concentration_ratio'])

# Add the columns
test_user_features = pd.concat([test_user_features, test_features_df], axis=1)
```

```python
active_users = train_user_features[train_user_features['history_length'] != 0].copy()
inactive_users = train_user_features[train_user_features['history_length'] == 0].copy()
print(len(inactive_users))
```

Note that we only do K-means on active-users, **not** the **cold-start users** as they could skew the clusters towards (0,0). Cold start users have a separate label, 'Cold Start'.

```python
user_level_diversity_df_train = train_user_features[['user_id', 'diversity_ratio', 'concentration_ratio', 'history_length']].drop_duplicates(subset=['user_id']).set_index('user_id')

active_users_unique = user_level_diversity_df_train[user_level_diversity_df_train['history_length'] != 0].copy()

features = active_users_unique[['diversity_ratio', 'concentration_ratio']]
scaler = StandardScaler()
# Fit and transform the numerical data
scaled_data = scaler.fit_transform(features)

# Store the scaled data features into a dataframe, preserving the user_id index
scaled_cluster_features = pd.DataFrame(scaled_data, columns=['diversity_ratio', 'concentration_ratio'], index=active_users_unique.index)
print(features)
```

```python
# Convert to numpy array to fit into KMeans algorithm
df_matrix = scaled_cluster_features.to_numpy()

# Choose optimal K using Elbow method
cost = []

# Loop across every possible cluster
for cluster in range(1, 10):
    kmeans = KMeans(n_clusters = cluster, random_state = 0)

    # Fit and predict each customer's cluster
    kmeans.fit_predict(df_matrix)

    # Store the cost, with the element's index being the cluster number
    cost.append(kmeans.inertia_)
    print('Cluster initiation: {}'.format(cluster))

    # Storing the cost of each cluster into a dataframe
df_cost = pd.DataFrame({'Cluster': range(1, 10), 'Cost':cost})
```

```python
# Plots the cost with respect to the number of clusters
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  # x series
  xs = series['Cluster']
  # y series
  ys = series['Cost']
  # plot
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')

# Sort by ascending order of cluster
df_sorted = df_cost.sort_values('Cluster', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Cluster')
_ = plt.ylabel('Cost')
```

##4.1 Knowledge Match Score

*   Cosine similarity between the user’s reading profile and a candidate news article using entity embeddings learned from a knowledge graph with TransE

*   Measures how well the article’s underlying entities and concepts align with the user’s knowledge interests



```python
def normalize_news_id(x):
    if x is None:
        return None
    if isinstance(x, float) and pd.isna(x):
        return None

    s = str(x).strip()
    if s == "":
        return None

    if s.startswith("N"):
        s = s[1:]

    try:
        f = float(s)
        if f.is_integer():
            s = str(int(f))
        else:
            s = str(f)
    except ValueError:
        pass

    return f"N{s}"


def normalize_history(x):
    if isinstance(x, list):
        return [normalize_news_id(i) for i in x if normalize_news_id(i) is not None]
    if isinstance(x, np.ndarray):
        return [normalize_news_id(i) for i in x.tolist() if normalize_news_id(i) is not None]
    if isinstance(x, str):
        return [normalize_news_id(i) for i in x.split() if normalize_news_id(i) is not None]
    return []


def ensure_list(x):
    if isinstance(x, list):
        return x
    if isinstance(x, np.ndarray):
        return x.tolist()
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return []
    return [x]


def safe_array(x):
    if isinstance(x, np.ndarray):
        return x.astype(np.float32)
    if isinstance(x, list):
        return np.array(x, dtype=np.float32)
    return None


def build_article_vectors(news_entities):
    kg_vectors = []
    semantic_vectors = []

    kg_dim = None
    sem_dim = None

    for _, row in news_entities.iterrows():
        kg_title = safe_array(row["kg_title_emb"])
        kg_abs = safe_array(row["kg_abstract_emb"])
        ner_title = safe_array(row["ner_title_emb"])
        ner_abs = safe_array(row["ner_abstract_emb"])

        if kg_dim is None:
            for v in [kg_title, kg_abs]:
                if v is not None:
                    kg_dim = v.shape[0]
                    break

        if sem_dim is None:
            for v in [ner_title, ner_abs]:
                if v is not None:
                    sem_dim = v.shape[0]
                    break

        kg_vecs = []
        if row["kg_title_flag"] == 1 and kg_title is not None:
            kg_vecs.append(kg_title)
        if row["kg_abstract_flag"] == 1 and kg_abs is not None:
            kg_vecs.append(kg_abs)

        if kg_vecs:
            kg_vectors.append(np.mean(kg_vecs, axis=0))
        else:
            kg_vectors.append(np.zeros(kg_dim, dtype=np.float32))

        sem_vecs = []
        if row["ner_title_flag"] == 1 and ner_title is not None:
            sem_vecs.append(ner_title)
        if row["ner_abstract_flag"] == 1 and ner_abs is not None:
            sem_vecs.append(ner_abs)

        if sem_vecs:
            semantic_vectors.append(np.mean(sem_vecs, axis=0))
        else:
            semantic_vectors.append(np.zeros(sem_dim, dtype=np.float32))

    kg_matrix = np.vstack(kg_vectors).astype(np.float32)
    semantic_matrix = np.vstack(semantic_vectors).astype(np.float32)

    id_to_index = {
        normalize_news_id(nid): i
        for i, nid in enumerate(news_entities["news_id"].tolist())
    }

    return kg_matrix, semantic_matrix, id_to_index


def cosine_sim(a, b):
    if a is None or b is None:
        return 0.0

    na = np.linalg.norm(a)
    nb = np.linalg.norm(b)

    if na == 0 or nb == 0:
        return 0.0

    return float(np.dot(a, b) / (na * nb))

def compute_knowledge_scores_grouped(
    df,
    kg_matrix,
    semantic_matrix,
    id_to_index,
    history_col="history"
):
    df = df.copy()

    candidate_news_all = []
    labels_all = []
    kg_scores_all = []
    sem_scores_all = []
    combined_scores_all = []

    kg_dim = kg_matrix.shape[1]
    sem_dim = semantic_matrix.shape[1]

    for _, row in df.iterrows():
        history = normalize_history(row[history_col])

        # build candidates in fixed order: positive first, then negatives
        if "pos_id" in df.columns and pd.notna(row.get("pos_id", None)):
            pos_candidates = [row["pos_id"]]
        elif "pos_ids" in df.columns:
            pos_candidates = ensure_list(row.get("pos_ids", []))
        else:
            pos_candidates = []

        if "neg_ids" in df.columns:
            neg_candidates = ensure_list(row.get("neg_ids", []))
        elif "neg_id" in df.columns and pd.notna(row.get("neg_id", None)):
            neg_candidates = [row["neg_id"]]
        else:
            neg_candidates = []

        pos_candidates = [normalize_news_id(x) for x in pos_candidates]
        pos_candidates = [x for x in pos_candidates if x is not None]

        neg_candidates = [normalize_news_id(x) for x in neg_candidates]
        neg_candidates = [x for x in neg_candidates if x is not None]

        candidate_news = pos_candidates + neg_candidates
        labels = [1] * len(pos_candidates) + [0] * len(neg_candidates)

        # user vector from history
        hist_idx = [id_to_index[h] for h in history if h in id_to_index]

        if hist_idx:
            user_kg = kg_matrix[hist_idx].mean(axis=0)
            user_sem = semantic_matrix[hist_idx].mean(axis=0)
        else:
            user_kg = np.zeros(kg_dim, dtype=np.float32)
            user_sem = np.zeros(sem_dim, dtype=np.float32)

        kg_scores = []
        sem_scores = []
        combined_scores = []

        for news_id in candidate_news:
            if news_id in id_to_index:
                idx = id_to_index[news_id]
                news_kg = kg_matrix[idx]
                news_sem = semantic_matrix[idx]
            else:
                news_kg = None
                news_sem = None

            kg_score = cosine_sim(user_kg, news_kg)
            sem_score = cosine_sim(user_sem, news_sem)
            combined_score = 0.5 * kg_score + 0.5 * sem_score

            kg_scores.append(kg_score)
            sem_scores.append(sem_score)
            combined_scores.append(combined_score)

        candidate_news_all.append(candidate_news)
        labels_all.append(labels)
        kg_scores_all.append(kg_scores)
        sem_scores_all.append(sem_scores)
        combined_scores_all.append(combined_scores)

    df["candidate_news"] = candidate_news_all
    df["labels"] = labels_all
    df["knowledge_match_score_kg"] = kg_scores_all
    df["knowledge_match_score_semantic"] = sem_scores_all
    df["knowledge_match_score_combined"] = combined_scores_all

    return df

kg_matrix, semantic_matrix, id_to_index = build_article_vectors(news_entities)

train_scored = compute_knowledge_scores_grouped(train, kg_matrix, semantic_matrix, id_to_index)
val_scored = compute_knowledge_scores_grouped(val, kg_matrix, semantic_matrix, id_to_index)
test_scored = compute_knowledge_scores_grouped(test, kg_matrix, semantic_matrix, id_to_index)

print(train_scored[[
    "impression_id",
    "candidate_news",
    "labels",
    "knowledge_match_score_combined"
]].head())

print(val_scored[[
    "impression_id",
    "candidate_news",
    "labels",
    "knowledge_match_score_combined"
]].head())

print(test_scored[[
    "impression_id",
    "candidate_news",
    "labels",
    "knowledge_match_score_combined"
]].head())
```

## 5.1 Hour of the Day
We previously observed from our EDA that news consumption in the MIND dataset is not uniform. User click-through-rate (CTR) tend to peak during specific windows, such as in the morning (7:00 AM – 9:00 AM) and evening hours (8:00 PM – 10:00 PM). This suggests that the time of day is a strong contextual predictor of a user's willingness to engage with news.

We also saw that certain categories experience higher CTR during specific hours, such as 'music-celebrity' news gaining popularity towards the later evening hours, possibly the period of leisurely scrolling. Hence we include `hour` as a contextual feature of the impression.

* We perform **cyclical transformation** on the hour. While the raw hour (0–23) is a numerical value, it is categorical and cyclical in nature.
> If we feed the raw integer into the model, it would treat the distance between 11:00 PM (Hour 23) and 12:00 AM (Hour 0) as 23 units, when in reality, they are only 1 hour apart.
* To resolve this, we map the 24-hour cycle onto a unit circle using Sine and Cosine transformations
> * Sine ($\sin$): Captures the vertical position on the clock face.
> * Cosine ($\cos$): Captures the horizontal position on the clock face.


```python
import numpy as np

train['hour'] = train['time'].dt.hour

train['hour_sin'] = np.sin(2 * np.pi * train['hour'] / 24)
train['hour_cos'] = np.cos(2 * np.pi * train['hour'] / 24)

train = train.drop(columns=['hour'])
```

## 5.2 Habitual Context of Impression
We introduce `habit_score` to provide user-personalised temporal context. This is as from the EDA, we observe that many users check the news at the same exact time every day, at 24-hour intervals.

The feature is derived by calculating the time interval in hours between the current impression and the user's most recent previous interaction. We then apply a periodic transformation:

$$\cos\left(\frac{2\pi \times \text{interval}}{24}\right)$$

* Score = 1.0 (Habitual): Occurs when the interval is a multiple of 24 hours (0, 24, 48, 72). This indicates the user has returned at the exact same time of day, suggesting a strong habitual return.

* Score = -1.0 (Off-Schedule): Occurs when the interval is 12, 36, or 60 hours. This indicates the user is logging in at the exact opposite time of their usual routine (e.g., a morning reader logging in at night).

* Score = 0.0 (Neutral/New): This is assigned to a user's first-ever visit in the dataset or during 6-hour/18-hour offsets. It represents a neutral state where the model should not rely on daily cycles to predict behavior.

```python
train = train.sort_values(by=['user_id', 'time'])

# Sort chronologically by user
train = train.sort_values(by=['user_id', 'time'])

# Shift to get the previous time for each user
train['prev_time'] = train.groupby('user_id')['time'].shift(1)

# Calculate the time difference in hours
train['interval_hours'] = (train['time'] - train['prev_time']).dt.total_seconds() / 3600

# Apply the cosine transformation (1.0 for 24h multiples, -1.0 for 12h offsets)
train['habit_score'] = np.cos(2 * np.pi * train['interval_hours'] / 24)

# Impute 0.0 for the first visit of every user (where interval is NaN)
train['habit_score'] = train['habit_score'].fillna(0.0)

# Drop temporary calculation columns
train = train.drop(columns=['prev_time', 'interval_hours'])
```

##5.3 Recency Score of News Article
We have the recency score of each candidate news article to be a context feature, as its value is determined by the relationship between the article and the specific impression the user encounters it.

* We calculate this by calcluating $: 1 / (1 + \text{time since first impression})$
* A very fresh article (~0 hours old) gets a score of $1.0$. An article that is 24 hours old gets a score of $1/25 \approx 0.04$. This allows the MLP to learn that relevance of an article decreases exponentially after the first few hours.


```python
print("Sample of the news dataset:")
display(news.head())
```

Recency score computation

```python
import random

def shuffle_candidates_numpy(pos_id, neg_ids):
    # 1. Escape NumPy: Convert to standard Python integer and list
    p_id  = int(pos_id)
    n_ids = [int(x) for x in neg_ids]

    # 2. Concatenate
    candidates = [p_id] + n_ids
    labels     = [1] + [0] * len(n_ids)

    # 3. Zip, shuffle, and unzip
    combined = list(zip(candidates, labels))
    random.shuffle(combined)
    shuffled_candidates, shuffled_labels = zip(*combined)

    return list(shuffled_candidates), list(shuffled_labels)
```

```python
train[['candidate_articles', 'labels']] = train.apply(
    lambda row: shuffle_candidates_numpy(row['pos_id'], row['neg_ids']),
    axis=1,
    result_type='expand'
)
```

```python
train.drop(columns=['pos_id', 'neg_ids'])
```

```python
release_time_map = dict(zip(news['news_id'], news['first_train_impression_time']))

def compute_recency_list(impression_time, candidate_articles):
    scores = []

    for news_id in candidate_articles:
        # Fetch the first impression time, defaulting to NaT if missing
        first_seen = release_time_map.get(news_id, pd.NaT)

        # Handle unseen articles (NaN)
        if pd.isna(first_seen):
            # If it has never been seen in train, it is completely fresh (age = 0)
            # Score = 1 / (1 + 0) = 1.0
            scores.append(1.0)
            continue

        # Calculate the age in hours
        age_hours = (impression_time - first_seen).total_seconds() / 3600.0

        # Safeguard against negative age (if impression_time < first_seen due to timestamp anomalies)
        age_hours = max(0, age_hours)

        # Apply the recency decay formula
        score = 1.0 / (1.0 + age_hours)
        scores.append(score)

    return scores

# 3. Apply using a fast list comprehension
train['recency_scores'] = [
    compute_recency_list(imp_time, candidates)
    for imp_time, candidates in zip(train['time'], train['candidate_articles'])
]
```
